```python
v1 = (4,3)
v2 = (-1,1)


def add(v1,v2):
    return (v1[0]+v2[0],v1[1]+v2[1])

total = add(v1,v2)
origin = (0,0)
draw(
    Points(*[v1,v2,total]),
    Arrow(v1,origin,color=red),
    Arrow(v2,origin,color=green),
    Arrow(total,origin,color=blue),
    Arrow(total,v1,color=green),
)
```
